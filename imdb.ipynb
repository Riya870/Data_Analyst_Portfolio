```python
list = []
for value in data['Genre']:
    list.append(value.split(','))
```

```python
list
```

```python
# convert 2D list into 1D list
one_d = []
for item in list:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
uni_list = []
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```
